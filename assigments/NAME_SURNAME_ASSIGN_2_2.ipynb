```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
```

```python
feature_names = iris.feature_names
features = pd.DataFrame(iris.data, columns = feature_names)
#target = pd.DataFrame(iris.target, columns = ['target'])
target = iris.target
target_names = iris.target_names

new_list = []
for i in target:
    if i == 0:
        new_list.append(target_names[0])
    elif i == 1:
        new_list.append(target_names[1])
    elif i == 2:
        new_list.append(target_names[2])
        
target = pd.DataFrame(new_list, columns = ['target'])

iris_dataframe = pd.concat([features, target],  axis = 1)
```

## Question 10
* mean center each column of the iris dataframe (excluding the label column of course)
* this means, for each column, the mean should be zero.  to accomplish this we can subtract the column mean from each element of the column
* note, broadcasting, which if we have say a 150 row and 4 column dataframe, can  take a row vector of size 4 and apply it to each element
*  thus, we can answer this questions doing something like df - df_col_means
* run the describe() method at the end to show the data has been mean center

```python
#get mean of each column
names_and_means = iris_dataframe.mean()
```

```python
#store name of each column in generator
col_names = (k for k in names_and_means.index)

#store column mean values in generator
col_mean = (k for k in names_and_means)
```

```python
for col_number in range(0,4):
    x = next(col_names)
    iris_dataframe[x] = iris_dataframe[x] - next(col_mean)
```
